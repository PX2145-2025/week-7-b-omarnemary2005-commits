# Introduction
Random processes are very common in physics. When dealing with large numbers of molecules, such as in gases, we donâ€™t describe the trajectory of every single particle ($10^{23}$ particles is too much to handle) but often use probabilities to determine the average behaviour. Here, we look at some illustrative examples of using random numbers in physical applications.

## Random walks
The random walk problem is a classic example which, despite its simplicity, describes very well the basic physical phenomenon of _diffusion_. Diffusion is a process where par- ticles which are initially concentrated in a region in space spread out with time. For example, a drop of ink in water will gradually spread to the whole volume of water, even if the water is perfectly still. This is because the individual molecules (both the ink and the water) (i) have random directions and speeds, which depend on the temperature, and (ii) also collide with each other. It turns out that these two ingredients can be modelled very easily.

## The method
We first look at a random walk in 1-D for a single particle. We start the particle off at the origin of the x-axis and assume that it moves unit distance in each time step (the molecular speed is thus 1 unit). At the end of each time step we set the direction to left or right with equal probability, mimicking the effect of collisions. You can easily write the python code, but for flexibility of later python coding, I show here how to define a function, called `walk(steps)`, which is passed a single number steps and returns a vector of that length, which contains all the positions of the random walk.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x
````

IMPORTANT: Make sure you understand how this function is used, and how it works. The rest of this worksheet is based on using it. You can also use the vectorised version below.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    Vectorised method.
    '''
    x = np.random.choice([-1, 1], size=steps) # random -1 or 1
    x = np.cumsum(x) # cumulative sum
    return x
````

Those two quantities, the "mean position" and the "mean squared position" are standard quantities in random walks.
````python
# part 1 generate and plot 10 random walks of 1000 steps

steps = 1000  #number of steps
N = 10  # number of walkers

# Creating a 2D array to store the positions where the rows represent the step number and the columns represent the random walk
particles = np.zeros((steps, N), int)

# Fill array with random walks
for w in range(N):    # Loop over each walker
    steps_array = np.random.choice([-1, 1], size=steps) # random steps of equal probability of either -1 or +1
    particles[:, w] = np.cumsum(steps_array)   # cumulative sum to get positions stored in column w of the particles array

# Plot all 10 walks
plt.figure()
for w in range(N):
    plt.plot(particles[:, w])
plt.xlabel("Step Number")
plt.ylabel("Position (x)")
plt.title("10 Random Walkers, 1000 steps each")
plt.show()
````

````python
# part 2
#Calculating the steps for 500 walkers and defining for each step a mean position ane a mean squared position
steps = 1000  # number of steps
N = 500   # number of walkers

# Create array to store all walkers
particles = np.zeros((steps, N), int)

# Generate each walker using the same method as before
for w in range(N):
    steps_array = np.random.choice([-1, 1], size=steps)  # random steps of equal probability of either -1 or +1
    particles[:, w] = np.cumsum(steps_array)  # cumulative sum to get positions stored in column w of the particles array

# Calculate the mean position âŸ¨xâŸ© at each step
mean_position = np.mean(particles, axis=1)

# Calculate the mean squared position âŸ¨xÂ²âŸ© at each step
mean_squared_position = np.mean(particles**2, axis=1)

# Print final values at the last step (step 1000)
print("Mean position at final step =", mean_position[-1])
print("Mean squared position at final step =", mean_squared_position[-1])
````

````python
# Parts 3 and 4 :Mean position and mean squared position plots with explanation

steps = 1000
N = 500   # number of walkers

# Create array to store all walkers
particles = np.zeros((steps, N), int)

# Generate each walker
for w in range(N):
    steps_array = np.random.choice([-1, 1], size=steps)  # random steps of equal probability of either -1 and 1
    particles[:, w] = np.cumsum(steps_array)             # cumulative sum to get positions stored in column w of the particles array

# Mean position âŸ¨xâŸ© at each step
mean_position = np.mean(particles, axis=1)

# Mean squared position âŸ¨xÂ²âŸ© at each step
mean_squared_position = np.mean(particles**2, axis=1)

# --- Part 3 Plot ---
plt.figure()
plt.plot(mean_position)
plt.xlabel("Step Number")
plt.ylabel("Mean Position âŸ¨xâŸ©")
plt.title("Mean Position vs Step (500 Walkers)")
plt.show()

# --- Part 4 Plot ---
plt.figure()
plt.plot(mean_squared_position)
plt.xlabel("Step Number")
plt.ylabel("Mean Squared Position âŸ¨xÂ²âŸ©")
plt.title("Mean Squared Position vs Step (500 Walkers)")
plt.show()
````

The mean position fluctuates around zero because the random walk is unbiased, meaning left and right steps are equally likely so they cancel out on average. However, the mean squared position increases steadily with the number of steps. This shows that the walkers spread out more over time, even though there is no overall drift in one direction. This increasing spread is a key characteristic of diffusion.
